## Task 1

Write a program that computes and prints the first $n$ rows of the Pascal’s triangle, where $n\geq 0$ is the row number with $n=0$ corresponding to the topmost row (which is a single element 1). Use your program to print all the rows up to $n=10$.

Enter your comments on how you solved the task here (in English):


```python
n=11      # here 11 is given because in question it asked me to print the rows from 0 to 10
L1=[]
for i in range(n):
    t_list = []                # nested list is created
    for j in range (i+1):         # i is row and j is column
        if j==0 or j==i:
            t_list.append(1)   # for storing first and last value as 1 
        else:
            t_list.append(L1[i-1][j-1] + L1[i-1][j])  # adds the previous two values and stores in the list
    L1.append(t_list)       # stores the list inside the list to form nested list as like forming matrix
#print(L1)

# now to print the list in a triangle shape the below code is used
for i in range(n):
    for j in range(n-i-1):    # n-i-1 is used to print the space before it
        print(format(" ", "<2"), end="")
    for j in range(i+1):      # to print how many numbers we want in each row
        print(format(L1[i][j],"<2"), end="  ")   #it is taken from the list where the number is stored.
    print()
```

Analyse your solution here (in English):
 
firstly the empty list is created and another list is created and appended the 1 in first and last index then middile value is created by adding the above two values. List is stored in a main list to form the nested list. 
Secondly forloop is created to print the list values in the form of triangle.
